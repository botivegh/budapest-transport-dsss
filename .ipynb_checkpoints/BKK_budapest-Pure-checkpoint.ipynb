```python
import pandas as pd
import numpy as np
```

```python
path = 'BKK_budapest_gtfs'
#Read the feed with gtfs-kit
#routes, stops, stop_times, trips, shapes

routes = pd.read_csv(path + "/routes.txt", delimiter=",", low_memory=True)
stops = pd.read_csv(path + "/stops.txt", delimiter=",", low_memory=True)
stop_times = pd.read_csv(path + "/stop_times.txt", delimiter=",", low_memory=True)
trips = pd.read_csv(path + "/trips.txt", delimiter=",", low_memory=True)
shapes = pd.read_csv(path + "/shapes.txt", delimiter=",", low_memory=True)
calendar_dates = pd.read_csv(path + "/calendar_dates.txt", delimiter=",", low_memory=True)

### Add dates
trips  =pd.merge(trips, calendar_dates, on="service_id", how='left') 

### Add route type and more to trips
trips = pd.merge(trips, routes[['route_id','route_short_name', 'route_desc', 'route_color', 'route_type']], how='left', on="route_id")
```

```python
del calendar_dates
del routes
```

### Time transform and Create hourly separation

```python
#stop_times['arrival_h'] = stop_times.arrival_time.apply(lambda x : int(x.split(":")[0]))
#stop_times['departure_h'] = stop_times.departure_time.apply(lambda x : int(x.split(":")[0]))
```

```python
stop_times.arrival_time = pd.to_datetime( stop_times.arrival_time, format= '%H:%M:%S', errors='coerce')
stop_times.departure_time = pd.to_datetime(stop_times.departure_time, format= '%H:%M:%S', errors='coerce')
```

```python
stop_times.dropna(subset=["departure_time"], inplace=True)
```

```python
### Check stops in every 10 mins activity in the months

stop_times.resample('10T', on='arrival_time').stop_id.count().plot.line()
```

```python
#stop_times.resample('H', on='arrival_time').stop_id.count().plot.line()
```

### Calcualte individual trips' statistics to use later for clustring

- avg. trip speed
- trip stop density by 1km
- total activity (number of times that same trip taken during a months)

```python
stop_times_trips = stop_times.groupby('trip_id').agg({'departure_time' : 'first' , 'arrival_time' : 'last', 'shape_dist_traveled' : 'max', 'stop_sequence' : 'max', 'stop_id' : 'count' })
```

```python
stop_times_trips.head(2)
```

```python
##Speed
stop_times_trips['duration'] = stop_times_trips.arrival_time -  stop_times_trips.departure_time
stop_times_trips['duration'] = stop_times_trips['duration'].dt.total_seconds() / 60 / 60
stop_times_trips['speed'] = stop_times_trips.shape_dist_traveled /1000 / stop_times_trips['duration']

stop_times_trips.speed.replace([np.inf, -np.inf], np.nan)
stop_times_trips.dropna(subset=['speed'], inplace=True, how="all")

stop_times_trips = stop_times_trips[stop_times_trips.speed!=np.inf]
```

```python
## Stop density: how many stop the route/trip has per 1km
stop_times_trips['stop_density'] = stop_times_trips['stop_sequence'] / stop_times_trips['shape_dist_traveled'] * 1000
```

```python
stop_times_trips
```

### Merge and grouby stats with shapes data

```python
### Adding route information: route type, description and short name
```

```python
shape_stats = pd.merge(trips, stop_times_trips, how="inner", on="trip_id").groupby("shape_id").agg({'trip_id' : 'count', 'service_id' : 'first' ,"duration" : "mean" , 'speed' : "mean",'stop_density': 'mean', "stop_sequence" : "mean", 'route_short_name' :'first', 'route_desc':'first', 'route_color':'first', 'route_type':'first' })
#shape_stats =  pd.merge( shape_stats , avg_waitingtime_per_shape, on="shape_id", how='left' ) 
```

### Merge and grouby stats with shapes data - considering weekend

```python
shape_stats_week_weekend = pd.merge(trips, stop_times_trips, how="inner", on="trip_id")
shape_stats_week_weekend['weekend'] = pd.to_datetime(shape_stats_week_weekend['date'], format='%Y%m%d').dt.weekday
shape_stats_week_weekend.weekend  = shape_stats_week_weekend.weekend >= 5
```

```python
shape_stats_week_weekend = shape_stats_week_weekend.groupby(["shape_id", 'weekend']).agg({'trip_id' : 'count', 'service_id' : 'first' ,"duration" : "mean" , 'speed' : "mean",'stop_density': 'mean', "stop_sequence" : "mean", 'route_short_name' :'first', 'route_desc':'first', 'route_color':'first', 'route_type':'first' })
shape_stats_week_weekend.reset_index(inplace=True)
```

```python
## trip_id = trip taken during the months period (no. vehicule on the route)
## stop_sequence = no. stop in the route
## arrival_time = avg waiting time on the route
shape_stats.rename(columns={ 'trip_id' : 'count_trips', 'stop_sequence' : 'count_stops'}, inplace=True)
shape_stats_week_weekend.rename(columns={ 'trip_id' : 'count_trips', 'stop_sequence' : 'count_stops'}, inplace=True)
```

```python
shape_stats['route_type'].value_counts()
```

# Analyse - Bus route clustering

```python
### Filter to Bus
shape_stats_bus = shape_stats[shape_stats.route_type==3]
```

### Standardise

```python
from sklearn.preprocessing import RobustScaler, MinMaxScaler
rs = RobustScaler(quantile_range=(10.0, 90.0))
minmax = MinMaxScaler()
```

```python
shape_stats_bus.count_trips.plot.hist(bins=50)
```

```python
normed = shape_stats_bus.copy()
```

```python
normed[['count_trips', 'count_stops', 'speed']] = rs.fit_transform(normed[['count_trips', 'stop_density', 'speed']])
```

```python
normed = normed[['count_trips', 'stop_density', 'speed']]
```

### Kmeans

### Elbow method

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# calculate SSE for a range of number of cluster
list_SSE = []
min_k = 1
max_k = 10
range_k = range(min_k, max_k+1)
for i in range_k:
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(normed)
    # inertia is a concept in physics. Roughly it means SSE of clustering.
    list_SSE.append(km.inertia_)

# plot
plt.plot(range_k, list_SSE, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
k_cluster = 4
random_seed = 1
kmeans_method = KMeans(n_clusters=k_cluster,random_state=random_seed)
kmeans_method.fit(normed)
```

```python
shape_stats_bus['KC cluster'] = kmeans_method.labels_
```

```python
shape_stats_bus.head(2)
```

```python
shape_stats_bus.to_csv('./output/bus_shape_stats.csv')
```

### Shapes to GeoPandas

```python
import contextily as ctx
import geopandas as gpd
from shapely.geometry import Point, LineString
```

```python
shape_stats =  pd.read_csv('./output/bus_shape_stats.csv')
```

```python
shapes['geometry'] = shapes.apply(lambda x : (x.shape_pt_lon, x.shape_pt_lat), axis=1)
gdf_shapes = gpd.GeoDataFrame(pd.DataFrame(shapes.groupby('shape_id').agg({ 'shape_dist_traveled' : 'max' ,'geometry' : lambda x : LineString(list(x))})), geometry='geometry', crs='epsg:4326')
```

```python
gdf_shapes = pd.merge(gdf_shapes, shape_stats, how="inner", on="shape_id")
```

```python
gdf_shapes['color'] = pd.qcut(gdf_shapes.speed, q=[0, .25, .5, .75, 1.], labels=['red', 'orange', 'yellow', 'green'])
```

```python
colors = {0 : 'red', 1: 'orange', 2:'yellow', 3:'green'}
gdf_shapes['color'] =  gdf_shapes['KC cluster'].apply(lambda x : colors[x])
```

```python
gdf_shapes.groupby('KC cluster').agg({'speed' : 'mean', 'count_trips' : 'mean',  'stop_density' : 'mean'})
```

```python
gdf_shapes[gdf_shapes['color']=='red']
```

```python
gdf_shapes = gdf_shapes.to_crs(epsg=3857)
ax = gdf_shapes.plot(figsize=(22, 22), alpha=0.3, color=gdf_shapes['color'] )
#gdf_shapes[gdf_shapes['shape_id']=="SW00"].plot(alpha=1,  color='blue', ax=ax)
#gdf_shapes[gdf_shapes['shape_id']=="X485"].plot(alpha=1,  color='green', ax=ax)
ctx.add_basemap(ax, zoom=12, source=ctx.providers.CartoDB.DarkMatterNoLabels)
```

## Traffic of stops

```python
stops = pd.merge(stops, stop_times.groupby('stop_id').agg({'trip_id' : 'count'}).sort_values('trip_id'), how="left", on="stop_id")
```

```python
stops.dropna( subset=['trip_id'], inplace=True)
```

```python
stops.rename(columns={'trip_id' : 'no_stops_in_months'}, inplace=True)
```

```python
stops = gpd.GeoDataFrame(stops, geometry=gpd.points_from_xy(stops.stop_lon,stops.stop_lat, crs='epsg:4326'))
```

```python
stops.to_file("./output/stops.geojson", driver="GeoJSON")
```
