### Visualise Bus Stops

```python
gdf_stops = gpd.GeoDataFrame(feed.stops, geometry=gpd.points_from_xy(feed.stops.stop_lon,feed.stops.stop_lat, crs='epsg:4326'))
gdf_stops = gdf_stops.to_crs(epsg=3857)
```

```python
ax = gdf_stops.plot(figsize=(10,10), markersize=1,color='blue', alpha=0.2)
ctx.add_basemap(ax, zoom=12, source=ctx.providers.Stamen.TonerLite)
```

### Create GeoPandas Dataframe out of the shapes (routes)

```python
feed.shapes['geom'] = feed.shapes.apply(lambda x : (x.shape_pt_lon, x.shape_pt_lat), axis=1)
```

```python
df = pd.DataFrame(feed.shapes.groupby('shape_id').agg({ 'shape_dist_traveled' : 'max', 'shape_id' : 'first' ,'geom' : lambda x : LineString(list(x))}))
```

```python
gdf_shapes = gpd.GeoDataFrame(df, geometry='geom', crs='epsg:4326')
```

```python
gdf_shapes = gdf_shapes.to_crs(epsg=3857)
```

```python
gdf_shapes[gdf_shapes['shape_id']=="SW00"]
```

```python
gdf_shapes[gdf_shapes['shape_id']=="X485"]
```

```python
ax = gdf_shapes.plot(figsize=(18, 10), alpha=0.02,  color='r')
gdf_shapes[gdf_shapes['shape_id']=="SW00"].plot(alpha=1,  color='blue', ax=ax)
gdf_shapes[gdf_shapes['shape_id']=="X485"].plot(alpha=1,  color='green', ax=ax)
#gdf_stops.plot(figsize=(10,10), markersize=1,color='blue', alpha=0.2,  ax=ax)
ctx.add_basemap(ax, zoom=12, source=ctx.providers.Stamen.TonerLite)
```
